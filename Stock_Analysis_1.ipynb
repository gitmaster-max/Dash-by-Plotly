```
import pandas as pd
import numpy as np
import pandas_datareader as pdr
import datetime as dt
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```
start = dt.datetime(2020, 1, 1)
end = dt.datetime(2020, 6, 1)

df = pdr.DataReader('RELIANCE.NS', 'yahoo', start, dt.datetime.now())
df.tail()
```

```
plt.figure(figsize=(15,8))
#plt.plot(df['Adj Close'] - df['Open'])
#plt.plot(df['High'] - df['Low'])
plt.plot(df['Adj Close'], linestyle='dashed', label='RELIANCE')
plt.title('RELIANCE ADJ CLOSE')
plt.xlabel('Month')
plt.ylabel('Adj close price')
plt.legend(loc='upper left')
```

```
SMA30 = pd.DataFrame()
SMA30['30 Day Moving Average'] = df['Adj Close'].rolling(window=30).mean()
SMA30
SMA120 = pd.DataFrame()
SMA120['30 Day Moving Average'] = df['Adj Close'].rolling(window=120).mean()
plt.figure(figsize=(15,8))
plt.plot(df['Adj Close'], label='RELIANCE')
plt.plot(SMA30, label='30 Day MA')
plt.plot(SMA120, label='120 Day MA')
plt.legend(loc='upper left')
```
